```python
import numpy as np
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d = pd.read_csv("ML Linear Regression Home Price Prediction & Analysis Project  Resource16932049360.csv")
d
```

```python
d.head()
```

```python
d.describe()
```

```python
d = d.drop(['Address', 'Date', 'Postcode', 'YearBuilt', 'Lattitude', 'Longtitude' ], axis=1)
d
```

```python
d.isnull().sum()
```

```python
d[['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']] = d[['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']].fillna(0)
d['Landsize'] = d['Landsize'].fillna(d.Landsize.mean())
d['BuildingArea'] = d['BuildingArea'].fillna(d.BuildingArea.mean())
```

```python
d.dropna(inplace=True)
```

```python
d.shape
```

```python
d.info()
```

```python
d['Method'].unique
```

```python
d = pd.get_dummies(d, drop_first=True)
#it helps in reducing the extra column created during dummy variable creation.
```

```python
d.head()
```

```python
x = d.drop('Price', axis=1)
y = d['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.3, random_state=2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
reg.score(x_train, y_train)
```

```python
reg.score(x_test, y_test)
```

```python
 from sklearn.linear_model import Ridge
```

```python
ridge_reg= Ridge(alpha=50, max_iter=100,)
ridge_reg.fit(x_train, y_train)
```

```python
ridge_reg.score(x_test, y_test)
```

```python
ridge_reg.score(x_train, y_train)
```

```python
from sklearn. linear_model import Lasso
```

```python
lasso_reg = Lasso(alpha=50, max_iter=100)
lasso_reg.fit(x_train, y_train)
```

```python
lasso_reg.score(x_test, y_test)
```

```python
lasso_reg.score(x_train, y_train)
```
